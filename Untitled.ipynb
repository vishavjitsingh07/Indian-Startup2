```python
def build_dict(array):
    d = {}
    #d holds investors name as key and the number of investments as value
    for i in array:
        if ',' not in i and i != "":
            d[i] = d.get(i, 0) + 1
        else:
            #splitting the Investors name list
            string = i.strip().split(',')
            for j in string:
                if j != "":
                    d[j.strip()] = d.get(j.strip(), 0) + 1
                
    return d

df.dropna(inplace = True, subset = ['InvestorsName'])
dict = build_dict(df['InvestorsName'])

invest = pd.DataFrame(list(dict.values()), list(dict.keys()))
a = invest.sort_values(by = [0], ascending = False)[:1]
print(a.index[0], a.values[0][0])
```
